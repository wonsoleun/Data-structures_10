16. 가능한 위상 정렬 순서는 4가지이다. a-b-c-d-e-f-h-g 혹은 a-d-e-b-c-f-h-g 혹은 a-b-d-c-e-f-h-g 혹은 a-d-b-e-c-f-h-g이다.

```python
create STACK
create visited <- false
for each vertex v
    if not visited[v]
        DFS(v, visited, STACK)
while not empty(STACK)
    pop(STACK)
DFS(v, visited, STACK)
  visited[v] <- true
  for i in Adj[v]
    if not visited[i]
        DFS
  STACK.push(v)
```
